```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

tips=sns.load_dataset('tips')
tips.head()
#setting a theme for plots
sns.set_theme() #grimsi bir arkaplan falan eklemiş oldu...
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',style='smoker',size='size',col='time') #hue ile smoker olanla olmayan farklı renklerle gösterilmiş olur
```

```python
#house price dataset
from sklearn.datasets import fetch_openml #sklearn.datasets ml projeleri için örnek dataset yüklememizi sağlar:)
housing = fetch_openml(name="house_prices", as_frame=True)
print(housing.data.head())
```

DISTRIBUTION PLOT

```python
sns.distplot(housing.data['LotFrontage']) #displot yazınca mavi renkli çizgiyi göstermez distplot yazınca gösterir..
```

CORRELATION
*   positive correlation
*  negative correlation

HEAT MAP

```python

correlation=housing.data.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,cmap='Blues',annot_kws={'size':8})
#cbar=True parametresiyle sağdaki renk çubuğu görünür hale gelir,square ile ısı harştası karelerden oluşmuş molur adı üstünde,annot=True dşyerek ısı haritasındaki karelerin
#üzerinde sayısal değerlerinin görünüp görünmeyeceğini belirlemiş oluruz,cmap parametresiyle renk belirlemiş oluruz mavi ve mavinin tonlarını belirlemiş olduk
#annot_kws={'size': 8}: Sayısal değerlerin yazı tipi boyutunu belirtir.
#fmt parametresiyle virgülden sonra ondalıklı kısmın kaç basamağının görüneceğini falan belirlemiş oluyoruz.cmapi reds yapaarak kırmızı tonlarına da ulaşabilirsin.
```
