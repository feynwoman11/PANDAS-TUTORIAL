```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

tips=sns.load_dataset('tips')
tips.head()
#setting a theme for plots
sns.set_theme() #grimsi bir arkaplan falan eklemi oldu...
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',style='smoker',size='size',col='time') #hue ile smoker olanla olmayan farkl renklerle gsterilmi olur
```

```python
#house price dataset
from sklearn.datasets import fetch_openml #sklearn.datasets ml projeleri iin rnek dataset yklememizi salar:)
housing = fetch_openml(name="house_prices", as_frame=True)
print(housing.data.head())
```

DISTRIBUTION PLOT

```python
sns.distplot(housing.data['LotFrontage']) #displot yaznca mavi renkli izgiyi gstermez distplot yaznca gsterir..
```

*   positive correlation
*  negative correlation

HEAT MAP

```python

correlation=housing.data.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,cmap='Blues',annot_kws={'size':8})
#cbar=True parametresiyle sadaki renk ubuu grnr hale gelir,square ile s hartas karelerden olumu molur ad stnde,annot=True dyerek s haritasndaki karelerin
#zerinde saysal deerlerinin grnp grnmeyeceini belirlemi oluruz,cmap parametresiyle renk belirlemi oluruz mavi ve mavinin tonlarn belirlemi olduk
#annot_kws={'size': 8}: Saysal deerlerin yaz tipi boyutunu belirtir.
#fmt parametresiyle virglden sonra ondalkl ksmn ka basamann grneceini falan belirlemi oluyoruz.cmapi reds yapaarak krmz tonlarna da ulaabilirsin.
```
